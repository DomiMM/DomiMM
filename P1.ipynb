```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for acetylcholinesterase
target = new_client.target
target_query = target.search('acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head(10)
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
df2
```

```python
df2 = df2.drop_duplicates(['canonical_smiles'])
```

```python
len(df2.canonical_smiles)
```

```python
len(df2.canonical_smiles.unique())
```

```python
# classify module in 3 catgeories
# active, inactive or intermediate
# active compund = drugs that have IC50 of less than 1 micromolar (or 1000 nano molar)
# inactive drug have a value between 1'000 and 10'000 nano molar

bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
len(bioactivity_class)
```

```python
selection = ["molecule_chembl_id", "canonical_smiles", "standard_value"]
df3 = df2[selection]
```

```python
df3
```

```python
df_4 = df3.reset_index()
```

```python
df_4
```

```python
df_5 = df_4.drop("index", axis=1)
```

```python
df_5
```

```python
df_6 = pd.concat([df_5,pd.Series(bioactivity_class)], axis=1)
```

```python
df_6
```

```python
df_7 = df_6.dropna()
```

```python
df_7
```

```python
df_7.to_csv('bioactivity_clean_and_processed_data.csv', index=False)
```
